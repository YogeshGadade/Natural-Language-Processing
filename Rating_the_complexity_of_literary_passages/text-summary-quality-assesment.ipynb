# About Project:
- The goal is to **assess the quality of summaries** written by students in grades 3-12 by predicting **content and wording scores** for each summary
- Project will assist teachers in evaluating the quality of student work and also help learning platforms provide immediate feedback to students.
- Created model will evaluate how well a student represents the main idea and details of a source text, as well as the clarity, precision, and fluency of the language used in the summary. 
- The possible challenge: evaluating summaries introduces an added layer of complexity, where models must consider both the student writing (summary) and a single, longer source text (prompt).
